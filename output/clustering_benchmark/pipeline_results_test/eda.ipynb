```python
import pandas as pd 

df = pd.read_csv(r"our_approach_summarization_results_test.csv")
```

```python
# Import necessary libraries
import pandas as pd
import ast  # For parsing strings into dictionaries


# Display the first few rows to understand the structure of the data
print("Data Preview:")
print(df.head())

# Define a function to parse the string representation of dictionaries
def parse_dict_column(column):
    return column.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Parse the columns containing dictionaries
df['kmeans'] = parse_dict_column(df['kmeans'])
df['agglomerative'] = parse_dict_column(df['agglomerative'])
df['dbscan'] = parse_dict_column(df['dbscan'])

# Extracting relevant metrics from each clustering method into separate DataFrames
methods = ['kmeans', 'agglomerative', 'dbscan']
all_metrics = ['num_clusters', 'mean_cluster_size', 'std_cluster_size', 'silhouette_score']

# Initialize an empty dictionary to hold the average scores for each method
average_scores = {method: {} for method in methods}

# Calculate the averages of each metric for each method
for method in methods:
    metrics = pd.json_normalize(df[method])  # Flatten the dictionary for each row into columns
    for metric in all_metrics:
        average_scores[method][f'average_{metric}'] = metrics[metric].mean()

# Convert the average scores dictionary into a DataFrame for better display
average_scores_df = pd.DataFrame.from_dict(average_scores, orient='index').reset_index()
average_scores_df = average_scores_df.rename(columns={'index': 'method'})

# Display the average scores for each method
print("\nAverage Scores for Clustering Methods:")
average_scores_df


```
